```python
# TRAINING,SCALING AND CREATING MODEL SECTION USING KERAS AND TENSORFLOW
```

```python
# TRAIN,SPLIT AND SCALING SECTION
```

```python
X = df.drop('price',axis=1).values  # we pass in values to make sure that there is
                                    # no datatypes conflicting hence we shall have  only numpy arrays
```

```python
y= df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=88)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train) # Here we are doing both fitting and transforming on the same line
                                        # We can achieve that using fit_transform(X_train) method
```

```python
X_test = scaler.transform(X_test)  # Here we only transform the test  data. We cant fit cos we
                                   # dont want to assume prior information about our test set
```

```python
X_train.max()  # testing the max of train dataset
```

```python
X_train.min()  #testing the min of the test dataset
```

```python
# CREATING THE MODEL
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(20,activation='relu'))   #the activation function is called rectified linear Unit
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')  #since this is a regression problem we are
                                             # choosing 'mse' as our loss metric

```

```python
# TRAINING THE MODEL
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
```

```python
model.history.history
```

```python
pd.DataFrame(model.history.history)  # You can put it in a dataframe
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()   #Then plot the losses
```

```python
# But the above model is overfitted due to the shape of the Orange line --> validation loss
```

```python
predictions = model.predict(X_test)
```
