### MachineLearning Section

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2,n_init=10)
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# Scatter plot of data points with colors based on cluster labels
plt.scatter(df['Private'], df['Private'], c=kmeans.labels_, cmap='viridis')

# Plot cluster centers as red crosses
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', s=100)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering Results')
```
