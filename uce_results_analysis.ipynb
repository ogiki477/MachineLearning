```python
import pandas as pd
import numpy as np 
import matplotlib as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
# df.columns  # to remove the whitespaces within the columns names
df.columns = df.columns.to_series().apply(lambda x: x.strip())
```

```python
df.columns
```

```python
# You can as well replace the whitespaces with underscores

df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df.info()
```

```python
# The areas of data cleaning
# 1. empty cells 
# 2. wrong format 
# 3. wrong data 
# 4. duplicates
```

```python
# 4. checking for duplicates and removing them 

for isduplicate in df.duplicated().array:
    if isduplicate:
        print("Duplictes found and dropped")
        df.drop_duplictes(inplace=True)

        break

    else:
        pass

print("No duplicates found")

df[df.duplicated()==True]
```

```python
# checking for null values 
df.isnull().sum().sort_values(ascending = True)
```

```python
# Checking null values interms of percentages
for i in df.columns:
    if df[i].isnull().any():
        print(i,"---",df[i].isnull().sum()*100/df.shape[0],"%")
```

```python
# CHecking for null values in figure 

import missingno as msno 
msno.bar(df,color='gray',fontsize=24)
```

```python
sns.heatmap(df.isnull())  # to visualize the missing values for the dataset
```

```python
# Checking the number of rows which contain null values 
withnullrows = df[df.isna().any(axis=1)]
beforeComputation = len(withnullrows.axes[0])

print('Number of rows with null values: ' + str(beforeComputation))
```

```python
# Removing rows without school names
df.dropna(thresh=None,subset=['SCHOOL'],inplace=True)
```

```python
# Replace missing TOTAL CANDIDATES with sum of male and female candidates since non of the  two have missing values
df["TOTAL_CANDIDATES"].fillna(df["MALE_CANDIDATES"] + df["FEMALE_CANDIDATES"],inplace=True)
```

```python
notEqual = df[df["TOTAL_CANDIDATES"] != (df["MALE_CANDIDATES"] + df["FEMALE_CANDIDATES"])]
notEqual[["SCHOOL","TOTAL_CANDIDATES","MALE_CANDIDATES","FEMALE_CANDIDATES"]]
```
