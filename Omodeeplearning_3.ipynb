```python
# PERFORMING A TRAIN TEST SPLIT
```

```python
X = df.drop('benign_0__mal_1',axis=1).values
```

```python
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
# SCALING THE DATA
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# CREATING THE MODEL
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape  # Right now in my training data i have 426 rows and 30 features
```

```python
# this is how you create the model
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

# since this is binary classfication,we shall specify activation as sigmoid for the ouput layer
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')

```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()    # the perfect example of overfitting  1. blue line is our training loss
                 # and the orange line is our validation loss
```

```python
# HOW TO DEAL WITH OVERFITTING
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

# since this is binary classfication,we shall specify activation as sigmoid for the ouput layer
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
# use of callbacks
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# help(EarlyStopping)  this code shows you more on how the EarlyStop callback works,u can run it
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=400,callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
# More on Reducing Overfitting
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

# since this is binary classfication,we shall specify activation as sigmoid for the ouput layer
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopped = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[early_stopped])
```

```python
Mymodel_loss = pd.DataFrame(model.history.history)
```

```python
Mymodel_loss.plot()
```

```python
# MORE ON REDUCING OVERFITTING WITH DROPOUT LAYERS
```

```python
# Another thing we could use to prevent overfitting is to add in dropout layers
# Dropout layers will turnoff a percentage of neuron layers randomly
```

```python
X_train.shape
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))



model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[early_stopped])
```

```python
Mymodel_loss1 = pd.DataFrame(model.history.history)
```

```python
Mymodel_loss1.plot()
```

```python
# FULL EVALUATION ON OUR CLASSES
# In the earlier versions of TensorFlow, the Sequential model had a predict_classes method.
# However, this method has been deprecated in newer versions,
# and the recommended approach is to use the predict method and apply np.argmax to obtain
# the predicted classes.

# As shown below

```

```python
predictions = np.argmax(model.predict(X_test),axis=1)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
